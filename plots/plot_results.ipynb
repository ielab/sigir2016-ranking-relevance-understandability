```python
%matplotlib inline
import pandas as pd
import numpy as np
import glob

import matplotlib
import matplotlib.pyplot as plt
```

```python
baseline_model="BM25" #"DirichletLM" #"TF_IDF", "BM25"
results_dir = "../terrier_scripts/results/"+baseline_model+"/*"
baseline_dir = "../terrier_scripts/baselines/*"
```

```python
def getBaselineDict(f):
    result = {}
    result["model"] = f.split("/")[-1]
    #result["goal"], result["feature_set"] = f.split("/")[-1].split("_")
    return getDict(f, result)
    
def getResultDict(f):
    result = {}
    result["model"] = f.split("/")[-2]
    result["goal"], result["feature_set"] = f.split("/")[-1].split("_")
    return getDict(f, result)
    
def getDict(f, result):
    fin = open(f, "r")
    for row in fin:
        fields = row.strip().split("\t")
        metric, value = fields[0].strip(), float(fields[2])
        result[metric] = value
    fin.close()
    return result
```

```python
files = glob.glob(results_dir)
df1 = pd.DataFrame.from_dict(getResultDict(files[0]), orient="index")
for f in files[1:]:
    df2 = pd.DataFrame.from_dict(getResultDict(f), orient="index")
    df1 = pd.concat((df1, df2), axis=1)
merged = df1.transpose().reset_index(drop=True)    

baselines_files = glob.glob(baseline_dir)
df1 = pd.DataFrame.from_dict(getBaselineDict(baselines_files[0]), orient="index")
for f in baselines_files[1:]:
    df2 = pd.DataFrame.from_dict(getBaselineDict(f), orient="index")
    df1 = pd.concat((df1, df2), axis=1)
baselines = df1.transpose().reset_index(drop=True)  
```

```python
def mysort(x):
    labels = ["binrel", "gradrel", "readOnly", "readTRel", "readP1TRel"]
    return labels.index(x)

def plot_graph(dataframe, baselines, baseline_model, metric, features_labels_y=0.255,
               graph_title="Precision at 10", ylimits=(0.25,0.38) ):
    ys = {}
    labels = {}
    sigs = {}
    
    for feature_set in ["ir", "read", "all"]:
        df_filter_feature = dataframe[dataframe["feature_set"] == feature_set]
        
        significance = list(df_filter_feature["SIG_" + metric].values)
        v = list(df_filter_feature[metric].values)
        k = list(df_filter_feature["goal"].values)
        k, v, significance = zip(*sorted(zip(k,v,significance), key=lambda x:mysort(x[0])))
        
        ys[feature_set] = v
        labels[feature_set] = k 
        sigs[feature_set] = significance
        #print v, k, significance

    bl_value = list(baselines[baselines["model"] == baseline_model][metric].values)

    n = sum([len(vs) for vs in ys.values()])
    xs = range(n)

    #Plot Baseline
    plt.plot(xs, len(xs)*bl_value)

    last_x = 0
    ticks_labels = []
    ticks_xs = []
    for feature_set in ["ir", "read", "all"]:
        xs = np.array(range(last_x, last_x + len(ys[feature_set])))
        last_x = last_x + len(ys[feature_set])
        
        a = np.array(sigs[feature_set])
        print xs, ys[feature_set], sigs[feature_set]
        p95 = a <= 0.05
        p90 = (a > 0.05) & (a <= 0.1)
        p85 = (a > 0.1) & (a <= 0.15)
        ns = (a > 0.15) 
        
        y = np.array(ys[feature_set])
        #return p95, xs, y
        # Significant:
        plt.scatter(xs[p95], y[p95], marker="D", s=50)
        plt.scatter(xs[p90], y[p90], marker="s", s=50)
        plt.scatter(xs[p85], y[p85], marker="^", s=50)
        plt.scatter(xs[ns], y[ns], s=50)
        
        
        ticks_labels.extend(labels[feature_set])
        ticks_xs.extend(xs)

    plt.xticks(ticks_xs, ticks_labels, rotation=45)
    
    plt.text(0.4, features_labels_y, 'IR\nFeatures')
    plt.plot([4.5,4.5],[-1.5,1.5], color="red")
    plt.text(5.4, features_labels_y, 'Readability\nFeatures')
    plt.plot([9.5,9.5],[-1.5,1.5], color="red")
    plt.text(10.4, features_labels_y, 'All\nFeatures')
    plt.ylim(ylimits)

    ax = plt.subplot(111)
    ax.grid(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["right"].set_visible(False)    
    plt.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")  

    plt.text(13, bl_value[0], 'Baseline')
    plt.xlim(-1,n)
    plt.title(graph_title, fontsize=14)
    outfilename = "%s_%s.pdf" % (baseline_model, metric.replace("(","").replace(")","").replace(".",""))
    print "Creating", outfilename
    #plt.savefig(outfilename, bbox_inches="tight", format='eps', dpi=1000)
    plt.savefig(outfilename, bbox_inches="tight")
```

```python
plot_graph(merged, baselines, baseline_model, "P_10", features_labels_y=0.255, graph_title="Precision at 10", ylimits=(0.25,0.38) )
```
